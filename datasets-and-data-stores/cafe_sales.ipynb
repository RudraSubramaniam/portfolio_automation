```python


df["Item"] = df["Item"].replace({"Missing value": pd.NA, "missing value": pd.NA, "UNKNOWN": pd.NA, "ERROR": pd.NA, "NaN": pd.NA, None: pd.NA}) 
df["Payment Method"] = df["Payment Method"].replace({"Missing value": pd.NA, "missing value": pd.NA, "UNKNOWN": pd.NA, "ERROR": pd.NA, "NaN": pd.NA, None: pd.NA}) 
df["Location"] = df["Location"].replace({"Missing value": pd.NA, "missing value": pd.NA, "UNKNOWN": pd.NA, "ERROR": pd.NA, "NaN": pd.NA, None: pd.NA}) 
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors="coerce")
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors="coerce")
```

```python
df["Quantity"] = df["Quantity"].astype("Int64")
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors="coerce")
```

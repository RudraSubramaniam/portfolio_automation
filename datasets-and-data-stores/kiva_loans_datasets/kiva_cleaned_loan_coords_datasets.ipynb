```python
import gc
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# 1. Define explicit file paths from the scratchpad directory
path_coords = r"C:\Users\sanjay\OneDrive\Desktop\scratchpad_archive\002.kiva_loans_datasets\loan_coords.csv"

# Load the source datasets into memory
df_coords = pd.read_csv(path_coords)
```

```python
print(df_coords.shape)
print(df_coords.columns)
print(df_coords.info())
df_coords.head(50)
```

```python
# Gather all structural metrics
missing_count = df_coords.isnull().sum()
missing_percent = (df_coords.isnull().sum() / len(df_coords)) * 100
column_types = df_coords.dtypes

# Combine them into a single DataFrame
full_report_coords = pd.DataFrame({
    "Data Type": column_types,
    "Missing Count": missing_count,
    "Missing %": missing_percent
})


# Note: We do not sort by "Missing %" here so that the columns 
# stay in their original order (0 to 102), making slicing accurate.

# Columns 1 to 25
print(full_report_coords.iloc[0:25])

# Strip hidden spaces from categorical string columns
coords_cols = ['loan_id', 'latitude', 'longitude']
```

```python
# Clean column headers
df_coords.columns = df_coords.columns.str.strip().str.lower()

# Check for missing coordinate rows
missing_lat = df_coords["latitude"].isna().sum()
missing_lng = df_coords["longitude"].isna().sum()
print(f"\nMissing Coordinates -> Lat: {missing_lat}, Lng: {missing_lng}")
```

```python
# Initialize a clean structural copy
df_coords_clean = df_coords.copy()

# Standardize column header strings
df_coords_clean.columns = df_coords_clean.columns.str.strip().str.lower()

# Deduplicate based strictly on the unique primary identifier
df_coords_clean = df_coords_clean.drop_duplicates(subset=['loan_id'])

# Save the fully preserved dataframe to an isolated staging file
df_coords_clean.to_csv(
    r"C:\Users\sanjay\OneDrive\Desktop\scratchpad_archive\002.kiva_loans_datasets\cleaned_loan_coords.csv",
    index=False
)

print("=== COORDS CLEANING VERIFICATION ===")
print(f"Verified Final Shape: {df_coords_clean.shape}")
print(df_coords_clean.isnull().sum())

print("Coords dataset clean state export successful!")
```

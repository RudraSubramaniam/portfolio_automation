```python
# 1. Define the collection of values you want to search for
target_strings = ['Y', 'No', 'yes', 'no']

# 2. Iterate through every column name in the DataFrame
for column_name in df.columns:
    
    # 3. Check if any row value matches the target strings list
    has_target_value = df[column_name].isin(target_strings).any()
    
    # 4. Print the column name if the condition evaluates to True
    if has_target_value:
        print(f"Target values found in column: {column_name}")
```

```python
# 1. Define the specific columns to investigate
target_columns = [
    'Street Suffix', 'Unit Suffix', 'Structural Notification', 
    'Voluntary Soft-Story Retrofit', 'Fire Only Permit', 
    'TIDF Compliance', 'Site Permit'
]

# 2. Establish the array of elements you want to ignore
ignore_values = ['Y', 'No', 'yes', 'no', 'Naa']

for col in target_columns:
    # 3. Build a condition to find values NOT in your ignore list AND NOT missing
    is_not_ignored = ~df[col].isin(ignore_values)
    is_not_missing = ~df[col].isna()
    
    # Combine both conditions using the bitwise AND operator
    unexpected_mask = is_not_ignored & is_not_missing
    
    # 4. Filter the dataframe and pull out only the remaining unique values
    unique_extras = df.loc[unexpected_mask, col].unique()
    
    print(f"Column [{col}] contains these unexpected values: {unique_extras}")
```

```python
# Convert the formatted as generic object strings explicitly to standard Pandas datetime formats:
date_cols = [
    "Permit Creation Date", "Current Status Date", "Filed Date", 
    "Issued Date", "Completed Date", "First Construction Document Date", 
    "Permit Expiration Date"
]
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], format="%m/%d/%Y", errors="coerce")
```

```python
# Text cleanup
df["Description"] = df["Description"].astype(str).str.strip().str.lower()

for col in ["Existing Use", "Proposed Use"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()
```

```python
# Text cleanup: astype(str) converts all data points in the column into string formatting to prevent errors from mixed types like integers or missing values. str.strip() removes any accidental leading or trailing spaces from the text boundaries. 
# str.lower() converts all characters in the column to lowercase format. str.title() capitalizes the first letter of every word and forces all other letters to lowercase format.
df["Description"] = df["Description"].astype(str).str.strip().str.lower()

for col in ["Existing Use", "Proposed Use"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()
```

```python
int_convert_cols = ["Unit", "Existing Units", "Proposed Units", "Plansets"]
for col in int_convert_cols:
    if col in df.columns:
        # Convert to numeric float first, round or drop decimals, then convert to Int64
        numeric_series = pd.to_numeric(df[col], errors="coerce")
        df[col] = numeric_series.round().astype("Int64")
```

```python
# Fix Zipcode format: 
if "Zipcode" in df.columns:
    df["Zipcode"] = df["Zipcode"].astype(str).str.split(".").str[0].str.strip()

# Replace empty strings with NaN
if "Unit_Clean" in df.columns:
    df["Unit_Clean"] = df["Unit_Clean"].replace(r"^\s*$", pd.NA, regex=True)
```

```python
# Create a copy dedicated for export formatting
df_export = df.copy()

# 1. Fill missing numerical structural fields to enforce strict integer output
int_cols = ["Unit", "Existing Units", "Proposed Units", "Plansets"]
for col in int_cols:
    if col in df_export.columns:
        df_export[col] = df_export[col].fillna(0).astype("int64")

# 2. Enforce strict string casting for mixed alphanumeric columns
if "Permit Number" in df_export.columns:
    df_export["Permit Number"] = df_export["Permit Number"].astype(str)

# 3. Format datetime values to clean date-only strings
date_cols = [
    "Permit Creation Date", "Current Status Date", "Filed Date", 
    "Issued Date", "Completed Date", "First Construction Document Date", 
    "Permit Expiration Date"
]
for col in date_cols:
    if col in df_export.columns:
        df_export[col] = df_export[col].dt.strftime("%Y-%m-%d").fillna("")

# 4. Save to a finalized distribution CSV
df_export.to_csv("san_francisco_building_permits_clean.csv", index=False)
```

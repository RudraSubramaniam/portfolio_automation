```python
import pandas as pd

# Load the dataset using Latin-1 encoding to handle special symbols like the British Pound sign
df = pd.read_csv(r"C:\Users\sanjay\OneDrive\Desktop\portfolio_github\001.Superstore_raw_dataset\Sample - Superstore.csv", encoding='Latin-1')
```

```python
print(df.shape)
print(df.columns)
print(df.info())
df.head(50)
```

```python
# Strip leading and trailing whitespaces from column names to prevent indexing errors
df.columns = df.columns.str.strip()

# Remove exact duplicate rows from the DataFrame to ensure transactional integrity
df = df.drop_duplicates()

# Convert the date columns using the correct matching format string
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%m/%d/%Y")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%m/%d/%Y")

# Convert Postal Code to string
df["Postal Code"] = df["Postal Code"].astype(str)
```

```python
from IPython.display import display, HTML

# 1. Gather all structural metrics
missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
column_types = df.dtypes

# 2. Combine them into a single DataFrame
full_report = pd.DataFrame({
    "Data Type": column_types,
    "Missing Count": missing_count,
    "Missing %": missing_percent
})

# Note: We do not sort by "Missing %" here so that the columns 
# stay in their original order (0 to 102), making slicing accurate.

# 3. Columns 1 to 25
print(full_report.iloc[0:25])

# Strip hidden spaces from categorical string columns
string_cols = ["Order ID", "Customer ID", "Customer Name", "Segment", "Country", "City", "State", "Ship Mode", "Region", "Product ID", "Category", "Sub-Category", "Product Name"]
for col in string_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
# Check for negative values in numeric columns
negative_sales = (df["Sales"] < 0).sum()
negative_qty = (df["Quantity"] < 0).sum()

# Check for logical date anomalies
date_anomalies = (df["Ship Date"] < df["Order Date"]).sum()

print(f"Negative Sales: {negative_sales}")
print(f"Negative Quantity: {negative_qty}")
print(f"Date Anomalies: {date_anomalies}")
```

```python
# Verify discounts are within valid percentage bounds
invalid_discounts = (df["Discount"] > 1.0).sum()
print(f"Invalid Discounts: {invalid_discounts}")
```

```python
# Create the formatting copy 
df_export = df.copy()

# Export to client-ready production file
df.to_csv("clean_superstore_data.csv", index=False)

print("Data export successful. Ready for import.")
```
